```python
import re
with open('emr_test.txt','r',encoding="utf-8") as f:
    cont = f.readlines()
output_data = []
item_f_list = []
for item in cont:
    item_list = item.strip().split(' ')
    for i in item_list:
        if i != '':
            item_f_list.append(i)
    for cont in item_f_list:
        if len(cont) ==1:
            output_data.append(cont+'\tS\n')
        else:
            output_data.append(cont[0]+'\tB\n')
            for w in cont[1:len(cont)-1]:
                output_data.append(w+'\tM\n')
            output_data.append(cont[len(cont)-1]+"\tE\n")
    output_data.append('\n')
result = ''.join(output_data)
print(result)
with open('emr_split_result.txt','w',encoding='utf-8') as f:
    f.write(result)
```

```python
# 计算一致性

# 按照格式返回人工分词文本内容
def openFile(fileName):
    with open(fileName,'r',encoding='utf-8') as f:
        cont = f.readlines()    
    item_f_list = []
    for item in cont:
        item_sub_list = []
        item_list = item.strip().split('。')
        for i in item_list:
            item_sub_list.append(i.split(' '))
        item_f_list.append(item_sub_list)
    return item_f_list

#  按格式返回自动分词文本
def openSegData(fileName):
    with open(fileName,'r',encoding='utf-8')as f:
        data = f.readlines()
    first_list = []
    for item in data:
        item_list = item.strip().split('。')
        second_list = []
        for cont in item_list:
            cont_list = jieba.cut(cont)
            cont_string = ' '.join(cont_list)
            third_list = cont_string.split(' ')
            #print(third_list)
            second_list.append(third_list)
        first_list.append(second_list)
    return first_list

#  计算一致性
def calConsis(file1,file2):
    file1_list = openFile('xu_seg.txt')
    file2_list = openSegData('seg_Data.txt')
    sent_num = 0.0
    sameItemNum = 0.0
    for num in range(len(file1_list)):
        for index in range(len(file1_list[num])):
            sent_num += 1
            flag = 0
            sameItem_list = []
            for item in file1_list[num][index]:
                if item in file2_list[num][index]:
                    flag += 1
                    sameItem_list.append(item)
            print('相同项数目:'+str(len(sameItem_list)))
            print('file1每句话分词数目: '+str(len(file1_list[num][index])))
            print('file2每句话分词数目：'+str(len(file2_list[num][index])))
            sameSubItemNum = len(sameItem_list)
            file1SubItemNum = len(file1_list[num][index])
            file2SubItemNum = len(file2_list[num][index])
            if sameSubItemNum == file2SubItemNum:
                sameItemNum += 1
            
    print('全部相同项数目：'+ str(sameItemNum))
    print('标准句子个数：'+str(sent_num))
    print('正确率：'+str(sameItemNum/sent_num))
    rate = sameItemNum/sent_num
    return rate
            
 
    
def main():
    rate = calConsis('xu_seg.txt','seg_Data.txt')
    print(rate)
                    
    
    
if __name__=='__main__':
    main()
```
